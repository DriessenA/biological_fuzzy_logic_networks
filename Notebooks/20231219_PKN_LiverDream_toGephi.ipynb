```python
import networkx as nx
from biological_fuzzy_logic_networks.DREAM import DREAMBioFuzzNet
```

```python
pkn_path = "/dccstor/ipc1/CAR/BFN/LiverDREAM_PKN.sif"
```

```python
### Save network for input in Gephi

G = DREAMBioFuzzNet.DREAMBioFuzzNet.build_DREAMBioFuzzNet_from_file(pkn_path)
for (n,d) in G.nodes(data=True):
    del d["output_state"]
    if "ground_truth" in d.keys():
            del d["ground_truth"]

for (e1,e2, d) in G.edges(data=True):
    if "layer" in d.keys():
        del d["layer"]

nx.write_gexf(G, "/dccstor/ipc1/CAR/BFN/LiverDREAM_PKN.gexf")
```
